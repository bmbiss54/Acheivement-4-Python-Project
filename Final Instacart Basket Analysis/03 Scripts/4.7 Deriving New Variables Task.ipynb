### If-Statements with User-Defined Functions

```python
# Developing Pricing Range Function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Testing and veiwing the price_label function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Checking the returns for the price range column
df['price_range'].value_counts(dropna= False)
```

```python
# checking higest price in data set
df['prices'].max()
```

### Step 4 Identify the busiest hours of the day

```python
# Checking order hour frequency
ords_prods_merged['order_hour'].value_counts(dropna = False)
```

There are 24 Hour sections in a day. We will separate these hours into three sections Most orders Average orders and Fewest orders.

```python
# creating loop for Most orders Average orders and Fewest orders"
result_hour = []

for value in ords_prods_merged["order_hour"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    result_hour.append("Most orders")
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
    result_hour.append("Fewest order")
  else:
    result_hour.append("Average Orders")
```

```python
# adding the column with result_hour to the dataframe
ords_prods_merged['busiest_period_of_day'] = result_hour
```

```python
# Checking to make sure the loop worked
ords_prods_merged[['order_hour','busiest_period_of_day']].head()
```

```python
# checking frequency of new busiest_period_of_day column
ords_prods_merged['busiest_period_of_day'].value_counts()
```

# Step 4 Creating a Loyalty flag using loc

```python
# developing a new column assigned to a gropu of aggregated order numbers by user id
ords_prods_merged['max_order'] = ords_prods_merged.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Removing row limiters
pd.options.display.max_rows = None
```

```python
ords_prods_merged.head(100)
```

```python
# Deriving loyalty flag after column creation
ords_prods_merged.loc[ords_prods_merged['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merged.loc[(ords_prods_merged['max_order'] <= 40) & (ords_prods_merged['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merged.loc[ords_prods_merged['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Viewing the loyalty flag frequencies and accuracies
ords_prods_merged['loyalty_flag'].value_counts(dropna = False)
```

```python
ords_prods_merged.head()
```

# Step 5 Checking Differences in Customer Loyalty Spending Habits

```python
# Grouping by loyalty flag and then running average price on those flags
ords_prods_merged.groupby('loyalty_flag').agg({'prices': ['mean','min','max']})
```

The purchasing bevaviours of the three groups is different based on average orders. The older loyalty customers spend the least per order, and the newer customers spend the most on their orders. You could infer the older customers have specialized what they buy on instacart and dont deviate on other products as much as a newer customer would. 

# Step 6 Creating Spending Flags For Customers

```python
# Need to group the data by user id and aggregate the average price by user
ords_prods_merged['average_price'] = ords_prods_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
pd.options.display.max_rows = None
```

```python
ords_prods_merged.head(100)
```

```python
# Need to develop flags of low spender and high spender based on new average price column made
ords_prods_merged.loc[ords_prods_merged['average_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merged.loc[ords_prods_merged['average_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Checking frequencies of the new spending flag results
ords_prods_merged['spending_flag'].value_counts(dropna = False)
```

# Step 7 Creating Frequency Flags for Customers

```python
# grouping data by user id and creating median day aggregation from days since prior order
ords_prods_merged['median_days'] = ords_prods_merged.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Check grouping
ords_prods_merged.head(100)
```

```python
# Creating flags for customer frequency types
ords_prods_merged.loc[ords_prods_merged['median_days'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merged.loc[(ords_prods_merged['median_days'] > 10) & (ords_prods_merged['median_days'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merged.loc[ords_prods_merged['median_days'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# Checking the new column frequencies and accuracies
ords_prods_merged['frequency_flag'].value_counts(dropna=False)
```
